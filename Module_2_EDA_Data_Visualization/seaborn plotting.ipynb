**Load dataset available in seaborn**

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white")

# Load mpg dataset
mpg = sns.load_dataset("mpg")
print(type(mpg))
```

```python
from tabulate import tabulate
import pandas as pd

print(tabulate(mpg, headers='keys', tablefmt='psql'))
```

**Scatterplot**

```python
sns.scatterplot(y ="mpg", x ="horsepower", data =mpg)
```

```python
# GRID is required for readability
sns.set(style="whitegrid")
sns.scatterplot(y ="mpg", x ="horsepower", data =mpg).set(title='mpg vs hp')
```

```python
# colour code with origin
sns.set(style="whitegrid")
sns.scatterplot(y ="mpg", x ="horsepower", hue='origin', data =mpg).set(title='mpg vs hp')
```

```python
# Points with diffret size based on nb of cylinders
sns.set(style="whitegrid")
sns.scatterplot(y ="mpg", x ="horsepower", hue='origin',size='cylinders', data =mpg).set(title='mpg vs hp')
```

```python
# with dark background and palette parameter
sns.set(style="whitegrid")
plt.style.use('dark_background')
sns.scatterplot(y ="mpg", x ="horsepower", hue='origin',size='cylinders',palette='bright', data =mpg).set(title='mpg vs hp')
```

**Facet Grid**

- Multi-Plot Grid for plotting conditional relationships

https://seaborn.pydata.org/generated/seaborn.FacetGrid.html



```python
# Load dataset

tips = sns.load_dataset("tips")
tips.head(20)
```

```python
# Plot facetgrid of scatter plots
sns.reset_defaults()
g = sns.FacetGrid(tips, col = 'time', row = 'smoker')
g.map(sns.scatterplot, "total_bill", "tip", facecolor='red')
```

```python
# Facetgrid with legend on scatterplot
g = sns.FacetGrid(tips, col = 'time', hue = 'smoker')
g.map(sns.scatterplot, "total_bill", "tip", edgecolor='white').add_legend()
```

```python
# Facetgrid with legend on scatterplot
g = sns.FacetGrid(tips, row = 'time', hue = 'smoker')
g.map(sns.scatterplot, "total_bill", "tip", edgecolor='white').add_legend()
```

```python
# Plot facetgrids of histogram
import numpy as np
sns.reset_defaults()
g = sns.FacetGrid(tips, row='sex', col='time')
g.map(sns.histplot, 'total_bill',bins=np.arange(0, 50, 5))
```

```python
# Facet Grid with lot of graphs
att = sns.load_dataset("attention")
print(tabulate(att, headers='keys', tablefmt='github'))
```

```python
g = sns.FacetGrid(att, col="subject", col_wrap=5,height=1.5)
g = g.map(sns.lineplot, "solutions", "score", marker=".")
```
