```python
# Angle between 2 vectors

np.random.seed(100)

v1 = np.random.rand(6)
v2 = np.random.rand(6)

print(v1)
print(v2)

angle = math.acos(np.dot(v1,v2)/(np.linalg.norm(v1,2) * np.linalg.norm(v2,2)))
print(angle)
```

```python
# Distance from a point to plane

import math
np.random.seed(100)

w = np.random.rand(6)
b = 1.2

x = np.random.rand(6)

d = abs(np.dot(w,x)+b)/np.linalg.norm(w,2)
print(d)
```

```python
# half-spaces
import math
np.random.seed(100)

w = np.random.rand(6)
b = 1.2

x = np.random.rand(6)

sign = np.sign(np.dot(w,x) + b / np.linalg.norm(w,2))
print(sign)
```

```python
# half-spaces test
import math
np.random.seed(100)

w = np.random.rand(6)
b = - 1.2

x = np.array([0,0,0,0,0,0])

sign = np.sign(np.dot(w,x) + b / np.linalg.norm(w,2))
print(sign)
```

```python
# Matrices operations

import numpy as np
import math

np.random.seed(100)

n=10
d=4

x1 = np.random.rand(n , d)
x2 = np.random.rand(n , d)

print(x1)
print('**********************************************************')
print(x2)

print('**********************************************************')
print(x1+x2)
print('**********************************************************')
print(x1*x2)
print('**********************************************************')
print(np.matmul(x1,x2.T))
```

```python
# Forbenius form

print(x1)

print(np.linalg.norm(x1,'fro'))

# implement using first principles

sh = x1.shape
print(sh)

sum=0
for i in range(sh[0]):
    for j in range(sh[1]):
        sum += x1[i,j]**2

print(math.sqrt(sum))
```

```python
# convolution
from scipy import signal
from scipy import misc

ascent = misc.ascent()
print(type(ascent))
print(ascent.shape)

import matplotlib.pyplot as plt
img_plot = plt.imshow(ascent, cmap='gray', vmin=0, vmax=255)
```

```python
k=20
kernel=np.ones((k,k))
kernel /= np.sum(kernel)

blurred = signal.convolve2d(ascent, kernel)
plt.imshow(blurred, cmap='gray', vmin=0, vmax=255)
```
