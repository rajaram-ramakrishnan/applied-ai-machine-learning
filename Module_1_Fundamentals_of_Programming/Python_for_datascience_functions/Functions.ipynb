```python
# Program to find HCF of 2 numbers

def computeHCF(num1, num2):
  '''
  This function computes the HCF between 2 numbers
  '''

  smaller = num2 if num1 > num2 else num1

  hcf = 1

  for i in range(1, smaller + 1):
    if (num1%i ==0) and (num2%i ==0):
        hcf=i
  
  return hcf
```

```python
num1 = 50
num2 = 49

print("HCF of {0} and {1} is {2}".format(num1, num2, computeHCF(num1, num2)))
```

```python
def computeHCF(num1, num2):
  '''
  This function computes the HCF between 2 numbers
  '''
  hcf = max( [i for i in range(1, min(num1,num2)+1) if (num1%i==0 and num2%i==0)])
  return hcf
```

```python
num1 = 49
num2 = 196

print("HCF of {0} and {1} is {2}".format(num1, num2, computeHCF(num1, num2)))
```

```python
def computeHCF(lstnums):

  '''
  This function computes the HCF of numbers in a list
  '''

  hcf = max(i for i in range(1, min(lstnums)+1) if all(j%i==0 for j in lstnums))
  return hcf
```

```python
import time
start = time.time()

hcf = computeHCF([33333333,66666666,99999999])

end = time.time()

print(hcf)
print('time taken :', end - start)
```

```python
def computeHCF(lstnums):

  '''
  This function computes the HCF of numbers in a list
  '''

  steps=2
  if all(j%2==0 for j in lstnums):
    steps=1

  hcf = max(i for i in range(1, min(lstnums)+steps) if all(j%i==0 for j in lstnums))
  return hcf
```

```python
import time
start = time.time()

hcf = computeHCF([33333333,66666666,99999999])

end = time.time()

print(hcf)
print('time taken :', end - start)
```

```python
# Python program for simple calculator

def add(x,y):
  '''
  This function adds two numbers
  '''

  return x + y

def subtract(x,y):
  '''
  This function subtracts two numbers
  '''

  return x - y

def multiply(x,y):
  '''
  This function multiplies two numbers
  '''

  return x *y

def divide(x,y):
  '''
  This function divides two numbers
  '''

  return x / y


print("Select Option")
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")

choice = int(input("Enter your choice 1/2/3/4 "))

if choice not in [1,2,3,4]:
  print("Invalid Choice")
else :
    num1 = int(input("Enter first number : "))
    num2 = int(input("Enter second number : "))

if choice == 1 :
  print("Addition of two numbers {0} and {1} is {2}".format(num1,num2,add(num1,num2)))
elif choice == 2 :
  print("Subtraction of two numbers {0} and {1} is {2}".format(num1,num2,subtract(num1,num2)))
elif choice == 3 :
  print("Multiplication of two numbers {0} and {1} is {2}".format(num1,num2,multiply(num1,num2)))
elif choice == 4 :
  print("Division of two numbers {0} and {1} is {2}".format(num1,num2,divide(num1,num2)))

```

```python
def factorial(num):
  '''
  This is a recursive function returns factorial of a number
  '''

  return 1 if num == 1 else (num* factorial(num-1))
```

```python
factorial(5)
```

```python
def HCFrecur(a, b):
  '''
  This function returns HCF of 2 numbers in a recursive way
  '''
  if b==0:
      HCF = a
  else:
    HCF = HCFrecur(b, a%b)
  
  return HCF
```

```python
HCFrecur(48,60)
```

```python
def fibonacci(count):
    fib_list = [0, 1]
  
    any(map(lambda _: fib_list.append(sum(fib_list[-2:])),
                                         range(2, count)))
  
    return fib_list[:count]
  
print(fibonacci(5))
```

```python
from functools import reduce

fib = lambda n: reduce(lambda x, _: x+[x[-1]+x[-2]],
								range(n-2), [0, 1])

print(fib(5))
```

```python
import sys

lst = ['b', 0 , 2]

for entry in lst:
  try:
    print("The entry is : ", entry)
    r = 1/int(entry)
  except(ValueError):
    print("This is value error")
  except(ZeroDivisionError):
    print("This is zero divide error")
  except:
    print("Other error")
  print("The reciprocal of", entry,"is", r)
```
