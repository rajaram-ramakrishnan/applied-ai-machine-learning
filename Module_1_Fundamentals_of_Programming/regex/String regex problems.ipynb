**Problem 1 : Mask personal information in email and phone numbers**

> Indented block



```python
# Mask email id
# email is example abcd@efg.com.in

txt = 'abcd@efg.comin'
x = re.search("^\w.*@\w.*[.]\w.*", txt)
print(x)
```

```python
# check email id is valid
import re
regex_verbose = re.compile(r"""
                           ^\w+([\.-]?\w+)*          # start
                           @                         # @ sign
                           \w+([\.-]?\w+)*           # domain
                           (\.\w{2,3})+$             # .com, .ac.in
                           """, re.VERBOSE | re.IGNORECASE)

res = regex_verbose.match("abcd.ab.cd.ef@iisc.ac.in")
print(res.string)
print(res)
```

```python
# PHONE NUMBER MASKING
#Example: 1(234)567-890 --> ###-###-7890"


ph = "1(234)567-890"

digits = re.sub("\D","", ph)
print(digits)

masked = "###-###-{}".format(digits[-4:])
print(masked)
```

```python
import PyPDF4
filepath = '/content/drive/MyDrive/Colab Notebooks/invoice.pdf'

with open(filepath,mode='rb') as f:
    reader = PyPDF4.PdfFileReader(f)
    page = reader.getPage(0)
    txt = page.extractText()
```

```python
# extract amount from invoices
amt = re.findall("\$[0-9]*\.[0-9]*", txt) # need to backspace for $
amt
```

```python
# extract total due
totaldue = re.findall('Total Due\n\$[0-9]*\.[0-9]*', txt)
print(totaldue[0])
totaldue = re.findall('\$[0-9]*\.[0-9]*', totaldue[0])
print(totaldue)
```

```python
# extract dates in the fixed format file
res = re.split('\n', txt)
print('Invoice date :', res[18])
```
