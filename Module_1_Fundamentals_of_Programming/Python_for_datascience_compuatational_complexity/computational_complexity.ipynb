```python
arr = [ 2, 11, 3, 78, 4, 47, 10, 57, 40, 99 ]
arr.sort()
print(arr)

search = 10

def binary_search(arr, low, high, x):
  print('********************************')
  print('low :', low)
  print('high :', high)

  if high >= low:
    mid = (high + low) //2
    print('mid value :', mid)
    print('Value at mid :', arr[mid])

    # Element is at midle
    if arr[mid] == x:
      return mid
    # Value is less than middle, only left side search
    elif arr[mid] > x :
      return binary_search(arr,low, mid -1, x)
    # value is more than middle, right side search
    else:
      return binary_search(arr,mid+1, high , x)
  else:
    return -1


result = binary_search(arr, 0, len(arr), search)
result
```

# **Find comon items in a list**

```python
l1 = list(range(100))
random.shuffle(l1)


l2 = list(range(50))
random.shuffle(l2)

cnt = 0
# complexity (m * n)
for i in l1:
  for j in l2:
    if i == j:
      print(i)
      cnt += 1

print("Number of common elements", cnt)
```

```python
l1 = list(range(100))
random.shuffle(l1)
print(len(l1))


l2 = list(range(50))
random.shuffle(l2)
print(len(l2))

# construct hashtable from big list

small = l2 if len(l2) < len(l1) else l1
big =   l1 if len(l2) < len(l1) else l2

# delete original lists
del l1
del l2

smalldict = {}
for ele in big:
  smalldict[ele] = 1;

cnt = 0
for i in small:
  if smalldict.get(i) != 'None':
    print(i)
    cnt += 1

print("Number of common elements", cnt)




```
