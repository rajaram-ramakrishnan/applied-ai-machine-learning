# **Shallow Copy examples**

## **list function**

```python
list1 = [1,2,3,4]
list2 = list(list1)
print(list2)
```

```python
list2.append(5)
list2
```

```python
list1
```

## **slicing operator**

```python
list1 = [1,2,3,4]
list2 = list1[:]
list2
```

```python
list2[0] = 'Raja'
list2
```

```python
list1
```

## **List Comprhension**

```python
list1 = [1,2,3,4]
list2 = [i for i in list1]
list2
```

```python
list2[3] = "a"
list2
```

```python
list1
```

## **Copy Function**

```python
import copy
list1 =[1,2,3,4]
list2 = copy.copy(list1)
list2
```

```python
list2.append("Hello")
list2
```

```python
list1
```

```python
list2[0] = 10
list2
```

```python
list1
```

## **Compound Objects**

```python
list1 = [[1,2],3,4]
list2 = copy.copy(list1)
list2
```

```python
list2.append("a")
list2
```

```python
list1
```

```python
list2[1]= 10
list2
```

```python
list1
```

```python
list2[0][1] = 20
list2
```

```python
list1
```
